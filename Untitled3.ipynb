```python
import torch, cv2, os, glob
import numpy as np
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Subset
import segmentation_models_pytorch as smp
from tqdm import tqdm
import albumentations as A
from albumentations.pytorch import ToTensorV2
from google.colab import drive

print("Сonnecting Google Drive")
drive.mount('/content/drive')

DEVICE = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print(f"Actuating device: {DEVICE}")

DATA_DIR = "/content/drive/MyDrive/full_dataset"
HOLD_DIR = os.path.join(DATA_DIR, "")
TEST_DIR = os.path.join(DATA_DIR, "test")

if not os.path.exists(HOLD_DIR):
    print(f"ERROR: Folder hold not found at path {HOLD_DIR}")
else:
    print(f"Found the hold folder in {HOLD_DIR}")

post_disaster_files = glob.glob(os.path.join(HOLD_DIR, "images", "*_post_disaster.png"))

print(post_disaster_files[-1])

ENCODER = "efficientnet-b3"
ENCODER_WEIGHTS = "imagenet"
N_CLASSES = 5  # 0=Background, 1=No damage, 2=Minor damage, 3=Moderate damage, 4=Destroyed
INPUT_CHANNELS = 6

BATCH_SIZE = 8
LEARNING_RATE = 0.0001
EPOCHS = 20
IMAGE_SIZE = 512
VALIDATION_SPLIT = 0.2
N = 2000


class BuildingDamageDataset(Dataset):
    def __init__(self, data_dir, image_size, augmentations=None, min_building_ratio=0.01):
        self.data_dir = data_dir
        self.image_size = image_size
        self.augmentations = augmentations
        self.valid_files = []

        post_disaster_files = glob.glob(os.path.join(self.data_dir, "images", "*post_disaster.png"))
        basename_list = [os.path.basename(file_path).replace("_post_disaster.png", "") for file_path in post_disaster_files]

        print(f"Filtering dataset (Threshold: {min_building_ratio*100}% buildings)")
        print(f"Scanning {len(basename_list)} files")

        skipped_empty = 0
        skipped_broken = 0

        for base_name in tqdm(basename_list, desc="Filtering"):
            pre_disaster_path = os.path.join(self.data_dir, "images", f"{base_name}_pre_disaster.png")
            post_disaster_path = os.path.join(self.data_dir, "images", f"{base_name}_post_disaster.png")
            target_path = os.path.join(self.data_dir, "masks", f"{base_name}_post_disaster.png")

            if not(os.path.exists(pre_disaster_path) and os.path.exists(post_disaster_path) and os.path.exists(target_path)):
                skipped_broken += 1
                continue

            else:
              self.valid_files.append(base_name)

        print(f"\nDone!")
        print(f"   Original: {len(basename_list)}")
        print(f"   Kept:     {len(self.valid_files)} (Useful)")
        print(f"   Removed:  {skipped_empty} (Too empty)")
        print(f"   Broken:   {skipped_broken}")

    def __len__(self):
        return len(self.valid_files)

      # Function for converting colors into classes (1, 2, 3, 4)
    def decode_mask(self, mask_rgb):

        mask_class = np.zeros(mask_rgb.shape[:2], dtype=np.uint8)

        # 1 No damage (Green)
        mask_class[np.all(mask_rgb == [1, 1, 1], axis=-1)] = 1

        # 2. Minor damage (Yellow)
        mask_class[np.all(mask_rgb == [2, 2, 2], axis=-1)] = 2

        # 3. Major damage (Orange)
        mask_class[np.all(mask_rgb == [3, 3, 3], axis=-1)] = 3

        # 4. Destroy (Red)
        mask_class[np.all(mask_rgb == [4, 4, 4], axis=-1)] = 4

        return mask_class

    def __getitem__(self, idx):
        base_name = self.valid_files[idx]

        pre_disaster_path = os.path.join(self.data_dir, "images", f"{base_name}_pre_disaster.png")
        post_disaster_path = os.path.join(self.data_dir, "images", f"{base_name}_post_disaster.png")
        target_path = os.path.join(self.data_dir, "masks", f"{base_name}_post_disaster.png")

        if not os.path.exists(pre_disaster_path):
             raise FileNotFoundError(f"File not found: {pre_disaster_path}")
        if not os.path.exists(post_disaster_path):
             raise FileNotFoundError(f"File not found: {post_disaster_path}")
        if not os.path.exists(target_path):
             raise FileNotFoundError(f"Mask not found: {target_path}")

        pre_image = cv2.imread(pre_disaster_path)
        post_image = cv2.imread(post_disaster_path)
        mask = cv2.imread(target_path)

        pre_image = cv2.cvtColor(pre_image, cv2.COLOR_BGR2RGB)
        post_image = cv2.cvtColor(post_image, cv2.COLOR_BGR2RGB)
        mask = self.decode_mask(mask)

        if pre_image is None or post_image is None:
            return self.__getitem__(np.random.randint(0, len(self.valid_files)))


        if self.augmentations:
            augmented = self.augmentations(image=pre_image, image1=post_image, mask=mask)
            pre_image = augmented['image']
            post_image = augmented['image1']
            mask = augmented['mask']

        pre_image = pre_image / 255.0
        post_image = post_image / 255.0
        to_tensor = ToTensorV2()
        processed = to_tensor(images=np.stack([pre_image, post_image], axis=0), mask=mask)

        return processed['images'][0].float(), processed['images'][1].float(), processed['mask'].long()

train_augs = A.Compose([
    A.RandomCrop(width=IMAGE_SIZE, height=IMAGE_SIZE),
    A.HorizontalFlip(p=0.5), A.VerticalFlip(p=0.5), A.RandomRotate90(p=0.5),
], additional_targets={'image1': 'image'})
val_augs = A.Compose([A.RandomCrop(width=IMAGE_SIZE, height=IMAGE_SIZE)], additional_targets={'image1': 'image'})


print("Creating datasets from the hold folder'")
dataset_for_train = BuildingDamageDataset(
    data_dir=HOLD_DIR,
    image_size=IMAGE_SIZE,
    augmentations=train_augs
)
dataset_for_val = BuildingDamageDataset(
    data_dir=HOLD_DIR,
    image_size=IMAGE_SIZE,
    augmentations=val_augs
)

dataset_size = len(dataset_for_train)
indices = list(range(dataset_size))
split = int(np.floor(VALIDATION_SPLIT * dataset_size)) # 20%
np.random.seed(42)
np.random.shuffle(indices)

train_indices, val_indices = indices[split:], indices[:split]

train_dataset = Subset(dataset_for_train, train_indices)
val_dataset = Subset(dataset_for_val, val_indices)

print(f"Size of the complete hold set: {dataset_size}")
print(f"Size of the training set (train): {len(train_dataset)}")
print(f"Size of the validation set (val): {len(val_dataset)}")

train_loader = DataLoader(
    train_dataset,
    batch_size=BATCH_SIZE,
    shuffle=True,
    num_workers=2
)

val_loader = DataLoader(
    val_dataset,
    batch_size=BATCH_SIZE,
    shuffle=False,
    num_workers=2
)

print(f"{len(train_loader)} batches have been created for training.")
print(f"{len(val_loader)} batches have been created for validation.")

print("Create model")

# U-Net model with EfficientNet-B3 encoder
building_detect_model = smp.Unet(
    encoder_name="efficientnet-b3",
    encoder_weights=None,
    in_channels=3, classes=1, activation='sigmoid'
)
building_detect_model.to(DEVICE)

damage_detect_model = smp.Unet(
    encoder_name="efficientnet-b3",
    encoder_weights="imagenet",
    in_channels=6,
    classes=5,
    activation=None
)

damage_detect_model.to(DEVICE)

weights = torch.tensor([0.1, 0.6, 1.3, 1.5, 1.4])

ce = torch.nn.CrossEntropyLoss(weight=weights.to(DEVICE))
dice = smp.losses.DiceLoss(mode="multiclass")

def loss_fn(logits, masks):
    return  0.5 * ce(logits, masks) + 0.5 * dice(logits, masks)

optimizer = optim.Adam(damage_detect_model.parameters(), lr=LEARNING_RATE)

print("The model, loss function, and optimizer are ready for training!")
```

```python
import time
import torch

SAVE_PATH_1 = "/content/drive/MyDrive/best_locator_model.pth"
SAVE_PATH_2 = "/content/drive/MyDrive/best_damage_model.pth"
if os.exists(SAVE_PATH_1):
  building_detect_model.load_state_dict(torch.load(SAVE_PATH_1))

if os.exists(SAVE_PATH_2):
  damage_detect_model.load_state_dict(torch.load(SAVE_PATH_2))


def validate_epoch(loader, model, loss_fn):
    model.eval()
    loop = tqdm(loader, desc="Validation")
    total_loss = 0.0

    with torch.no_grad():
        for pre_images, post_images, masks in loop:
            pre_images = pre_images.to(DEVICE)
            post_images = post_images.to(DEVICE)
            masks = masks.to(DEVICE)

            outputs = building_detect_model(pre_images)
            #pred_prob = outputs.squeeze()
            pred_masks = (outputs > 0.5).to(torch.uint8).unsqueeze(1)

            pre_images = pre_images * pred_masks
            post_images = post_images * pred_masks

            combined_images = torch.cat([pre_images, post_images], dim=2)

            outputs = damage_detect_model(combined_images)
            loss = loss_fn(outputs, masks)
            total_loss += loss.item()


    return total_loss / len(loader)

EPOCHS = 10

print(f"\n We are beginning training at {EPOCHS} eras....")

best_val_iou = 0.0

for epoch in range(EPOCHS):
    start_time = time.time()
    print(f"\n--- Era {epoch + 1}/{EPOCHS} ---")


    val_loss = validate_epoch(val_loader, model, loss_fn)

    duration = time.time() - start_time

    print(f"Time: {duration:.1f} secound")
    print(f"Val Loss:   {val_loss:.4f} (verification error)")

print("\n Training complete")
print(f"Best result IoU: {best_val_iou:.4f}")
```
